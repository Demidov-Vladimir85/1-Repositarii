## Проверка, что эффективность обогащения рассчитана правильно.

Вычислим её на обучающей выборке для признака 'rougher.output.recovery'. Что означает: эффективность обогощения на выходе после процесса флотации. Согласно описанию технологического процесса: данный коэффицент будет состоять из следующих величин:  
C = rougher.output.concentrate_au  
F = 'rougher.input.feed_ag'  
T = rougher.output.tail_au.   
Изначально определим кол-во пропусков, в перечеслинных выше столбцах. 

Выделем срез данных со значениями в столбце 'rougher.output.recovery' не равными NaN. Далее проверим кол-во пропусков в интересующих нас столбцах (выше приведен список из 4 столбцов) в полученном датафрейме.

```python
print('Кол-во пропусков:')
train[~train['rougher.output.recovery'].isnull()][['rougher.output.recovery', 'rougher.output.concentrate_au', 
                     'rougher.input.feed_ag', 'rougher.output.tail_au']].isnull().sum()
```

Для того, чтобы не создавать лишние глобальные переменные, произведем расчет через функцию. 

```python
def MAE_check(df):
    """
    Input parameters:
    df - dataframe
    """
    data_slice = df[~df['rougher.output.recovery'].isnull()]
    C = data_slice['rougher.output.concentrate_au']
    F = data_slice['rougher.input.feed_au']
    T = data_slice['rougher.output.tail_au']
    data_slice['recovery_calculated'] = (C * (F - T)) / (F * (C - T)) * 100
    MAE = mean_absolute_error(data_slice['recovery_calculated'], data_slice['rougher.output.recovery'])
    return MAE
```

```python
print(f"MAE состовляет: {MAE_check(train)}")
```

По графикам мы видим, что у нас есть выбросы в показаниях. У нас есть хвосты в низу распределений. Дальше возможны 2 сценария удаления выбросов:  
1) Определить по каждому из столбцов 1.5 межквартальных размаха и удалить выбросы.  
2) По каждому из столбцов определить 1.5 межквартальных размаха и найти пересечения индексов данных столбцов и удалить данные индексы.   

Так же необходимо заметить, что если сделать срез данных по какому-либо однмоу металлу с параметром концентрации равным нулю на одном из технологичеких этапов (например: rougher.output.concentrate_ag), то можно заметить, что концентрации по остальным металлам так же будут равны нулю. Исходя из той информации, которая у нас есть неясно, является ли это артефактом или это нормально с точки зрения технологического процесса.  

Далее будем придерживаться стратегии номер 2:  
Определить по каждому из столбцов 1.5 межквартальных размаха и удалить выбросы.  

Введем 2 функции:  
1) Функция iqr, которая будет считать 1.5 квартальных разамах.  
2) Фнкцция slice_preparation_wo_outliners, которая будет по каждому из столбцов датафрейма определять через функцию iqr интервалы выбросов и затем удалять эти значения. Размах интересующего нас интервала можно выбирать самостоятельно.

```python
def iqr(df, column):
    iqr = df[column].describe()[6] - df[column].describe()[4]
    low_bond = df[column].describe()[4] - 1.5 * iqr # считаем нижнюю границу
    if low_bond < 0: # бывает так, что нижняя граница уходит в минус, значит ей можно присвоить минимум признака 
        low_bond = df[column].describe()[3]
    upper_bond = df[column].describe()[6] + 1.5 * iqr # верхняя гранциа
    print('Нижняя граница:', low_bond)
    print('Верхняя граница:', upper_bond) 
```

```python
def slice_preparation_wo_outliners(df):
    final_array = []
    print(len(df))
    for column in df.columns[1:]:
        iqr = df[column].describe()[6] - df[column].describe()[4]
        low_bond = df[column].describe()[4] - 3.0 * iqr # считаем нижнюю границу 
        if low_bond < 0: # бывает так, что нижняя граница уходит в минус, значит ей можно присвоить минимум признака
            low_bond = df[column].describe()[3]
        upper_bond = df[column].describe()[6] + 3.0 * iqr # верхняя гранциа
        final_array.append([low_bond, upper_bond])
    for i in range(1, len(df.columns)):     
        df = df[(df.iloc[:, i] > final_array[i-1][0]) & (df.iloc[:,i] < final_array[i-1][1])] 
        #df = df[df.iloc[:, i] > 1.0]                
        print(len(df))
    return df
```

## Функция для вычесления итоговой sMAPE.

Напишем функцию sMAPE по которой будем считать эффективность моделей.

```python
def sMAPE(target, prediction):
    return np.sum(np.abs(target - prediction)/((np.abs(target) + np.abs(prediction)) / 2)) * 100/len(target)
```

```python
def final_sMAPE(floatt, final):
    final_sMAPE = 0
    final_sMAPE = 0.25 * floatt + 0.75 * final
    return final_sMAPE
```

```python
data_models = [
    
['Линейная Регрессия', 'sklearn.linear_model', 'LinearRegression', {}],
    
['Дерево решений в регрессии', 'sklearn.tree', 'DecisionTreeRegressor', 
{'max_depth': range(1, 10, 1), 'random_state': [12345], 'min_samples_leaf': range(2, 6), 'min_samples_split': range(2, 6)}],

['Случайный лес в регрессии','sklearn.ensemble', 'RandomForestRegressor', 
{'n_estimators': range(1, 5, 1), 'max_depth': range(1, 5, 1), 'random_state': [12345], 'min_samples_leaf': range(2, 6), 
             'min_samples_split': range(2, 6)}], 
    
['Модель Elastic Net','sklearn.linear_model', 'ElasticNet', 
{'l1_ratio': np.linspace(0, 1, 15, endpoint=True), 'alpha': np.linspace(0.1, 1, 10, endpoint=True), 'random_state': [12345]}]
    
              ] 
   
models_df = pd.DataFrame(data_models, columns = ['label','module','model', 'parameters'])
display(models_df)
```

```python
def grid_model_test_output(param_table, slice_fea_train_1, slice_tar_train_1, slice_fea_train_2, slice_tar_train_2):
    """
    Function for the best model definition based on our task's score sMAPE. The function returns a table with all models which
    user wants to check. The table contatins information on sMAPE, sMAPE delta vs constant model (dummy Regression), best model 
    parameters and best model (already fitted)
    
    Entry parameters:
    param_table - df with all necessary models and model parameters, listed through vocabulary
    slice_fea_train_1 - features train for target 1
    slice_tar_train_1 - train target 1
    slice_fea_train_2 - features train for target 2
    slice_tar_train_2 - train target 2
    
    Output parameters:
    result_df - dataframe with models info after GridSearchCV operation
    best_model_for_target_1 - best model, based on sMAPE, for target 1. Already fitted
    best_model_for_target_2 - best model, based on sMAPE, for target 2. Already fitted
    """
    the_best_score_result = final_sMAPE_DR
    result_df = []
    for index, row in param_table.iterrows():
        
        module_imported = importlib.import_module(row['module'])
        model_imported = getattr(module_imported, row['model'])
        model = model_imported()
        if verbose:
            print(model)
            
        grid_1 = GridSearchCV(model, row['parameters'], scoring = make_scorer(sMAPE, greater_is_better = False), 
                            cv = 5, verbose = 1, n_jobs = -1).fit(slice_fea_train_1, slice_tar_train_1)
        smape_rougher = sMAPE(test_rougher_output_recovery_target, 
                              grid_1.best_estimator_.predict(test_wo_outliners_for_target_one))
        
        grid_2 = GridSearchCV(model, row['parameters'], scoring = make_scorer(sMAPE, greater_is_better = False),
                             cv = 5, verbose =1, n_jobs = -1).fit(slice_fea_train_2, slice_tar_train_2)
        smape_clean = sMAPE(test_final_output_recovery_target, 
                              grid_2.best_estimator_.predict(test_wo_outliners_for_target_two))     
        
    
        
        final_sMAPE_result = final_sMAPE(smape_rougher, smape_clean)
        delta_vs_baseline = final_sMAPE_DR - final_sMAPE_result
        
        if final_sMAPE_result < the_best_score_result:
            the_best_score_result = final_sMAPE_result
            best_model_for_target_1 = grid_1.best_estimator_
            best_model_for_target_2 = grid_2.best_estimator_

             
        result_df.append([smape_rougher, smape_clean, final_sMAPE_result, row['label'], grid_1.best_params_, grid_2.best_params_, delta_vs_baseline])
        
    result_df = pd.DataFrame(result_df, columns = ['smape_rougher', 'smape_clean', 'Final sMAPE', 'Название модели', 'Оптимальные параметры_модель_1',
                                                  'Оптимальные параметры_модель_2', 'Дельта с константной моделью']) 
    
    result_df = result_df.sort_values(by = 'Final sMAPE', ascending = True)
    return result_df, best_model_for_target_1, best_model_for_target_2
```
