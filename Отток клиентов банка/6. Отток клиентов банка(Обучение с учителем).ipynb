### Подготовка данных к составлению моделей

```python
df = df.drop(['Surname', 'RowNumber', 'CustomerId'], axis = 1)
display(df.head())#проверка
df['Geography'].unique()# показывает что всего три страны
```

Получается 3 категориальных столбца, в одном 2 уникальных значения (пол), в другом всего 3 (страны). Для дальнейшего обучения любыми методами принимаем, что нужны только числовые значения. Для этих двух столбцов хорошо подходит метод One-Hot Encoding (значения не упорядоченные, малочисленные). Ввиду впринципе ненужности столбца с фамилиями для решения поставленной задачи этот столбец  вообще просто удаляем, как и столбцы RowNumber, CustomerId. Тем самим технически упрощаем задачу, снижаем нагрузку на производительность и спокойно можем применить метод One-Hot Encoding.

```python
df_ohe = pd.get_dummies(df, drop_first=True)# применяем метод прямого кодировани и удаляем первый столбец, чтобы избежать Дамми-ловушки

target = df_ohe['Exited']
features = df_ohe.drop('Exited', axis=1)

#encoder = OrdinalEncoder()
#df_ordinal = pd.DataFrame(encoder.fit_transform(df),
                            #columns=df.columns)

#target = df_ordinal['Exited']
#features = df_ordinal.drop('Exited', axis=1)

features_prom, features_test, target_prom, target_test = train_test_split(# разделение на тестовую и промежуточную выборки
    features, target, test_size=0.2, random_state=10000)

features_train, features_valid, target_train, target_valid = train_test_split(# разделение на обучающую и валидационную выборки
    features_prom, target_prom, test_size=0.25, random_state=10000)

print(features_train.shape, features_valid.shape, features_test.shape)#проверка разбивки 

numeric = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']
scaler = StandardScaler()
scaler.fit(features_train[numeric])

pd.options.mode.chained_assignment = None

features_train[numeric] = scaler.transform(features_train[numeric])
features_valid[numeric] = scaler.transform(features_valid[numeric])
features_test[numeric] = scaler.transform(features_test[numeric])
```

Произведена разбивка на тестовую, валидационную и обучающие выборки. 

```python
print((df_ohe['Exited'] == 1).sum() / (df_ohe['Exited'] == 0).sum())# проверка баланса целевого признака
```

Проверка баланса показала что его нет)) Соотношение 1 к 4. Далее обучаем модели не исправляя это недочет.

### Модель логистической регрессии

```python
model_log = LogisticRegression(random_state=10000, solver='liblinear')# инициация модели логистической регрессии 
model_log.fit(features_train, target_train) # обучение модели на тренировочной выборке
predicted_log = model_log.predict(features_valid)

print("F-1 мера модели логистической регрессии на валидационной выборке:", f1_score(target_valid, predicted_log))
```

### Модель решающего дерева

```python
best_model_tree = None
best_result_tree = 0
for depth in range(1, 10):
	model_tree = DecisionTreeClassifier(random_state=10000, max_depth=depth) # обучение модели с заданной глубиной дерева
	model_tree.fit(features_train, target_train) # обучение модели
	predicted_tree = model_tree.predict(features_valid) # предсказания модели
	result_tree = f1_score(target_valid, predicted_tree) #  качество модели
	if result_tree > best_result_tree:
		best_model_tree = model_tree
		best_result_tree = result_tree
print("F-1 мера модели решающего дерева на валидационной выборке:", best_result_tree)
```

### Модель случайного леса

```python
best_model_est = None
best_result_est = 0
for est in range(1, 30):
    model_est = RandomForestClassifier(random_state=10000, n_estimators=est) # обучение модели с заданным количеством деревьев
    model_est.fit(features_train, target_train) # обучение модели на тренировочной выборке
    predicted_est = model_est.predict(features_valid) # предсказания модели
    result_est = f1_score(target_valid, predicted_est) #  качество модели #  качество модели на валидационной выборке
    if result_est > best_result_est:
        best_model_est = model_est#  наилучшая модель
        best_result_est = result_est#   наилучшее значение метрики accuracy на валидационных данных
print("F-1 мера модели случайного леса на валидационной выборке:", best_result_est)        
        
```

### Выводы: 
По результатам исследования на трех моделях, наилучший показатель F-1 меры (даже немного превышающий значение 0,59, хоть это пока и не тест) показала модель решающего дерева. Модель случайного леса тоже показала высокий результат F-1 меры - 0,57 . А вот у логистической регрессии самый слабый результат 0,005. Вообще не подходит данная 
модель для такой задачи. Далее устраним дисбаланс и посмотрим на результат

Это значит что, логистическая регрессия совсем не может работать с несбалансированными моделями, но все равно имеет самое низкое значение F-1 меры.

### Метод увеличения выборки

```python
def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=10000)
    
    return features_upsampled, target_upsampled

features_upsampled, target_upsampled = upsample(features_train, target_train, 4)# 4 используем так как у нас такое соотношение
```

#### Модель логистической регрессии

```python
model_log_up = LogisticRegression(random_state=10000, solver='liblinear')
model_log_up.fit(features_upsampled, target_upsampled)
predicted_log_up = model_log_up.predict(features_valid)
print("F-1 мера модели логистической регрессии на валидационной выборке:", f1_score(target_valid, predicted_log_up))
```

####  Модель решающего дерева

```python
model_tree_up = DecisionTreeClassifier(random_state=10000, max_depth=7) # обучение модели с заданной глубиной дерева
model_tree_up.fit(features_upsampled, target_upsampled) # обучение модели
predicted_tree_up = model_tree_up.predict(features_valid) # предсказания модели
print("F-1 мера модели решающего дерева на валидационной выборке:",f1_score(target_valid, predicted_tree_up)) #  качество модели
```

#### Модель случайного леса

```python
model_est_up = RandomForestClassifier(random_state=10000, n_estimators=25) # обучение модели с заданным количеством деревьев
model_est_up.fit(features_upsampled, target_upsampled) # обучение модели на тренировочной выборке
predicted_est_up = model_est_up.predict(features_valid) # предсказания модели
print("F-1 мера модели случайного леса на валидационной выборке:", f1_score(target_valid, predicted_est_up)) #  качество модели
```

В принципе промежуточные выводы такие же, значительное увеличение показателя логистической регресии и незначительные колебания
показателей решающего дерева и случайного леса

### Метода уменьшения выборки

```python
def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_downsampled = pd.concat(
        [features_zeros.sample(frac=fraction, random_state=10000)] + [features_ones])
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=fraction, random_state=10000)] + [target_ones])
    
    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=10000)
    
    return features_downsampled, target_downsampled

features_downsampled, target_downsampled = downsample(features_train, target_train, 0.2561)
```

#### Модель логистической регрессии

```python
model_log_down = LogisticRegression(random_state=10000, solver='liblinear')
model_log_down.fit(features_downsampled, target_downsampled)
predicted_log_down = model_log_down.predict(features_valid)
print("F-1 мера модели логистической регрессии на валидационной выборке:", f1_score(target_valid, predicted_log_down))
```

#### Модель решающего дерева

```python
model_tree_down = DecisionTreeClassifier(random_state=10000, max_depth=7) # обучение модели с заданной глубиной дерева
model_tree_down.fit(features_downsampled, target_downsampled) # обучение модели
predicted_tree_down = model_tree_down.predict(features_valid) # предсказания модели
print("F-1 мера модели решающего дерева на валидационной выборке:",f1_score(target_valid, predicted_tree_down)) #  качество модели
```

#### Модель случайного леса

```python
model_est_down = RandomForestClassifier(random_state=10000, n_estimators=26) # обучение модели с заданным количеством деревьев
model_est_down.fit(features_downsampled, target_downsampled) # обучение модели на тренировочной выборке
predicted_est_down = model_est_down.predict(features_valid) # предсказания модели
print("F-1 мера модели случайного леса на валидационной выборке:", f1_score(target_valid, predicted_est_down)) #  качество модели
```

Получается для моделей решающего дерева и случайного леса результаты немного ухудшились, а для логистической модели немного увеличились. 

### Выводы 
Для данной задачи логистическая регрессия самый плохой вариант. Логистическая регрессия не работает с несбалансированными данными. Методы решающего дерева и случайного леса похожи, тенденции проявляются схоже. Метод решающего дерева чуть получше.
